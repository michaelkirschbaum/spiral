```python
import matplotlib.pyplot as plt
from math import sin, cos, radians, log, e

def draw_spiral(size):   
    # theta
    t = 420
    # constant
    a = 1
    
    '''
    calculate b
    size * cos(t) = acos(t)e^bt
    size = ae^bt
    size / a = e^bt
    ln(size / a) = ln(e^bt)
    ln(size / a) = bt * ln(e)
    ln(size / a) = bt
    ln(size / a) / t = b
    '''
    b = log(size / a) / radians(t)
    
    # distance from origin = ae^bt
    x = [a * (e ** (b * radians(t))) * cos(radians(t)) for t in range(720)]
    y = [a * (e ** (b * radians(t))) * sin(radians(t)) for t in range(720)]

    # resize
    plt.figure(figsize=(7, 7))
    
    # plt.xlim()
    # plt.ylim()
    
    # show x axis
    # plt.axhline(y=0, xmin=0, xmax=1)
    
    # show y axis
    # plt.axvline(x=0, ymin=0, ymax=1)

    # plt.grid(True)
    
    plt.plot(x, y)
    
    plt.show()
    
draw_spiral(50)
draw_spiral(30)
draw_spiral(10)
```
