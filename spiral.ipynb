```python
import matplotlib.pyplot as plt
from math import sin, cos, radians, log, e

def draw_spiral(size):   
    # rcos(t) = acos(t)e^bt
    
    # theta
    t = 60
    a = 1
    
    '''
    calculate b
    size * cos(t) = acos(t)e^bt
    size = ae^bt
    size / a = e^bt
    ln(size / a) = ln(e^bt)
    ln(size / a) = bt * ln(e)
    ln(size / a) = bt
    ln(size / a) / t = b
    '''
    b = log(size / a) / radians(t)
    
    # calculate distance for all radians
    # r = ae^bt
    x = [a * (e ** b * radians(t)) * cos(radians(t)) for t in range(720)]
    y = [a * (e ** b * radians(t)) * sin(radians(t)) for t in range(720)]

    # plt.figure(figsize=(7, 7))
    plt.xlim()
    plt.ylim()
    plt.axhline(y=0, xmin=0, xmax=1)
    plt.axvline(x=0, ymin=0, ymax=1)
    
    plt.plot(x, y)
    
    plt.show()
    
draw_spiral(50)
```
