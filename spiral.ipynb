```python
import matplotlib.pyplot as plt
from math import sin, cos, radians, log, e

def draw_spiral(size):   
    # theta
    t = 420
    # constant
    a = 1
    # revolutions about origin
    revs = 2
    
    '''
    calculate b
    size * cos(t) = acos(t)e^bt
    size = ae^bt
    size / a = e^bt
    ln(size / a) = ln(e^bt)
    ln(size / a) = bt * ln(e)
    ln(size / a) = bt
    ln(size / a) / t = b
    '''
    b = log(size / a) / radians(t)
    
    # distance from origin = ae^bt
    # distance * cos/sin = coordinate
    x = [a * (e ** (b * radians(t))) * cos(radians(t)) for t in range(revs * 360)]
    y = [a * (e ** (b * radians(t))) * sin(radians(t)) for t in range(revs * 360)]

    # resize
    # plt.figure(figsize=(7, 7))
    
    # fix window size
    plt.xlim(-800, 800)
    plt.ylim(-400, 400)
    
    # show x axis
    plt.axhline(y=0, xmin=0, xmax=1)
    
    # show y axis
    plt.axvline(x=0, ymin=0, ymax=1)

    # plt.grid(True)
    
    plt.plot(x, y)
    
    plt.show()

def main():
    draw_spiral(40)
    draw_spiral(45)
    draw_spiral(50)
    
if __name__ == "__main__":
    main()
```
